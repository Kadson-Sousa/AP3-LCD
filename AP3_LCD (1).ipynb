# Dataset

O DataSet utilizado neste trabalho foi obtido na base de dados abertos da Agência Nacional do Petróleo, Gás Natural e Biocombustíveis (ANP), órgão regulador do setor de combustíveis no Brasil.

Link para o DataSet: https://www.gov.br/anp/pt-br/centrais-de-conteudo/dados-abertos/serie-historica-de-precos-de-combustiveis

A base de dados é composta por registros de preços praticados em postos revendedores de todo o Brasil. Cada linha do dataset representa uma coleta individual em um estabelecimento específico, detalhando variáveis estruturais como o tipo de produto, a localização geográfica (município, estado e região) e o valor final de revenda.

O período analisado compreende os anos de 2019 a 2025, totalizando sete anos de série histórica. Os dados são disponibilizados semestralmente, portanto cada ano é carregado em dois arquivos distintos (1º e 2º semestre), que são concatenados em seguida.

A escolha desse recorte temporal foi motivada pelo interesse em analisar o comportamento dos preços de combustíveis em três momentos distintos: antes, durante e depois da pandemia de COVID-19, permitindo observar o impacto desse evento sobre os valores praticados e a posterior trajetória de recuperação do mercado no Brasil.

##Marcação de Outiliers

Antes de analisar a série temporal, verificamos a presença de valores atípicos no dataset completo utilizando o método IQR. Registros fora dos limites são considerados outliers.


```python
def marcar_outliers_iqr(grupo, coluna='Valor de Venda'):
    Q1 = grupo[coluna].quantile(0.25)
    Q3 = grupo[coluna].quantile(0.75)
    IQR = Q3 - Q1
    limite_inf = Q1 - 1.5 * IQR
    limite_sup = Q3 + 1.5 * IQR
    return (grupo[coluna] < limite_inf) | (grupo[coluna] > limite_sup)

#Filtrando e Removendo os Outliers do produto gasolina
df_filtrado_gasolina['Outlier_IQR'] = (
    df_filtrado_gasolina
    .groupby(['Estado - Sigla', 'AnoMes'])
    .apply(lambda g: marcar_outliers_iqr(g))
    .reset_index(level=[0,1], drop=True)
)

df_gasolina_sem_outliers = df_filtrado_gasolina[df_filtrado_gasolina['Outlier_IQR'] == False]

#Filtrando e Removendo os Outliers do produto etanol
df_filtrado_etanol['Outlier_IQR'] = (
    df_filtrado_etanol
    .groupby(['Estado - Sigla', 'AnoMes'])
    .apply(lambda g: marcar_outliers_iqr(g))
    .reset_index(level=[0,1], drop=True)
)

df_etanol_sem_outliers = df_filtrado_etanol[df_filtrado_etanol['Outlier_IQR'] == False]

#Filtrando e Removendo os Outliers do produto diesel
df_filtrado_diesel['Outlier_IQR'] = (
    df_filtrado_diesel
    .groupby(['Estado - Sigla', 'AnoMes'])
    .apply(lambda g: marcar_outliers_iqr(g))
    .reset_index(level=[0,1], drop=True)
)

df_diesel_sem_outliers = df_filtrado_diesel[df_filtrado_diesel['Outlier_IQR'] == False]
```

```python
# Quantidade de Outliers em cada produto
print("Gasolina:", df_filtrado_gasolina['Outlier_IQR'].sum())
print("Etanol:", df_filtrado_etanol['Outlier_IQR'].sum())
print("Diesel:", df_filtrado_diesel['Outlier_IQR'].sum())
```

### **Motivos da alta dos preços (2019-2022)**

No início de 2020, os preços da gasolina, do etanol e do diesel apresentaram redução em decorrência da queda na demanda provocada pelas medidas de isolamento social adotadas durante a pandemia de COVID-19. Com a retomada da atividade econômica, os preços passaram a registrar sucessivas altas.

**Gasolina**

A gasolina teve sua alta impulsionada pela valorização do petróleo no mercado internacional. Como a Petrobras adotava a política de paridade internacional de preços (PPI), as oscilações do mercado externo eram repassadas ao mercado interno. Em 2022, a guerra entre Rússia e Ucrânia intensificou esse movimento, levando a Petrobras a reajustar a gasolina em 18,8% em março daquele ano (PODER360, 2022). Em maio, o preço médio atingiu recorde histórico segundo a ANP (CNN BRASIL, 2022)

**Etanol**

O etanol subiu por dois motivos combinados. Primeiro, a safra de cana: a entressafra no início de 2021 fez o preço subir 21,1% em poucos meses, mais que a gasolina no mesmo período (MÁXIMO, 2021). Segundo, a relação de competitividade com a gasolina: como o etanol só compensa no motor flex quando custa até 70% do preço da gasolina, cada alta da gasolina abre espaço para o etanol subir junto sem perder competitividade, e as distribuidoras aproveitam essa margem pra reajustar o preço (MÁXIMO, 2021). Depois, uma seca voltou a prejudicar a produção, encarecendo também o álcool usado na mistura da gasolina (CNN BRASIL, 2021).

**Diesel S10**

O diesel foi o que mais subiu proporcionalmente. Em 2021 sozinho, o preço subiu 45% (FECOMBUSTÍVEIS, 2023). Em março de 2022, o reajuste da Petrobras também foi maior no diesel (24,9%) do que na gasolina (18,8%) (PODER360, 2022), chegando ao pico em junho daquele ano.

### **Motivo da queda após 2022**

A queda observada no segundo semestre de 2022 esteve associada às mudanças na tributação dos combustíveis, que alteraram a forma de cobrança do ICMS e contribuíram para reduzir os preços ao consumidor (SENADO FEDERAL, 2022).
